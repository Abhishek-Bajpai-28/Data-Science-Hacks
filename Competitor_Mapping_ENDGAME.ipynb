```python
##find correlation between modelling SKUs of Nestle and all competitors
mycorr_mf =[]
for mkt in markets:
        temp_df = df[(df['Channel']== mkt)]
        temp_df_nestle = temp_df[((temp_df['Manufacturer']== 'WAGNER') & (temp_df['Model_Flag']==1))]
        temp_df_comp = temp_df[temp_df['Manufacturer']!= 'WAGNER']
        Nestle_SKUs = temp_df_nestle.SKU.unique()
        Competition_SKUs = temp_df_comp.SKU.unique()
        for i in Nestle_SKUs:
            for j in Competition_SKUs:
                Array1 = temp_df_nestle['Volume'][(temp_df_nestle['SKU'] ==i)].reset_index(drop = True)
                Array2 = temp_df_comp['Avg_Price'][(temp_df_comp['SKU'] ==j)].reset_index(drop = True)
                xy = pd.DataFrame({'Vol1': Array1, 'Price2': Array2}).fillna(0)
                m = xy['Vol1'].corr(xy['Price2'])
                mycorr_mf.append(np.array([i,j,m]))
```
